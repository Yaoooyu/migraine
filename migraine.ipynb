```python
import pandas as pd
import os

file_path = '/root/autodl-fs/Migraine/20260130_dothepreprocessdata/migraine_with_id.xlsx'

if not os.path.exists(file_path):
    print(f"错误：找不到文件 {file_path}，请检查路径是否正确。")
else:
    
    try:
        df = pd.read_excel(file_path)
        print("✅ 文件读取成功！")
        
        print("\n--- 前 5 行数据 ---")
        display(df.head()) 

        print("\n--- 数据基本信息 (Info) ---")
        print(df.info())

        print("\n--- 数据统计描述 (Describe) ---")
        display(df.describe())
        
        print("\n--- 各列缺失值数量 ---")
        print(df.isnull().sum())
        
        print(f"\n--- 重复行数量: {df.duplicated().sum()} ---")

    except Exception as e:
        print(f"读取文件时发生错误: {e}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularDataset, TabularPredictor


if 'id' in df.columns:
    data = df.drop(columns=['id'])
else:
    data = df.copy()


train_data, test_data = train_test_split(
    data, 
    test_size=0.2, 
    random_state=42, 
    stratify=data['Migraine']
)


train_data = TabularDataset(train_data)
test_data = TabularDataset(test_data)

print(f"训练集大小: {train_data.shape}, 测试集大小: {test_data.shape}")



predictor = TabularPredictor(
    label='Migraine', 
    problem_type='binary',
    eval_metric='roc_auc',  
    path='AutogluonModels/ag_MPQ090_v1'
).fit(
    train_data,
    presets='medium_quality', # 'best_quality' 效果最好但慢，'medium_quality' 性价比高
    time_limit=300            
)


leaderboard = predictor.leaderboard(test_data, silent=True)

print("\n------ 模型排行榜 (Top 5) ------")
print(leaderboard[['model', 'score_test', 'score_val']].head())

performance = predictor.evaluate(test_data)
print("\n------ 最佳模型测试集表现 ------")
print(performance)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

best_model = predictor.get_model_best()
positive_class = predictor.class_labels[1] 

print(f"正在计算模型 {best_model} 的最佳阈值...")

y_true = test_data[predictor.label]
y_scores = predictor.predict_proba(test_data, model=best_model, as_pandas=True)[positive_class]

precisions, recalls, thresholds = precision_recall_curve(y_true, y_scores, pos_label=positive_class)


with np.errstate(divide='ignore', invalid='ignore'):
    f1_scores = 2 * (precisions * recalls) / (precisions + recalls)
    f1_scores = np.nan_to_num(f1_scores) 

f1_scores = f1_scores[:-1]


best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
max_f1 = f1_scores[best_idx]


print(f"最佳结果 ({best_model}) ------")
print(f"默认阈值F1: {f1_scores[np.abs(thresholds - 0.5).argmin()]:.4f}")
print(f"最佳阈值 (Threshold): {best_threshold:.4f}")
print(f"最佳 F1 Score:{max_f1:.4f}")


plt.figure(figsize=(10, 6))
plt.plot(thresholds, f1_scores, label='F1 Score Curve', color='tab:blue')


plt.scatter(best_threshold, max_f1, color='red', s=100, label=f'Optimal: {best_threshold:.2f}', zorder=5)
plt.axvline(best_threshold, color='red', linestyle='--', alpha=0.5)
plt.axhline(max_f1, color='red', linestyle='--', alpha=0.5)

plt.title(f'F1 Score vs Decision Threshold ({best_model})', fontsize=14)
plt.xlabel('Threshold', fontsize=12)
plt.ylabel('F1 Score', fontsize=12)
plt.legend(loc='best')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score


y_true = test_data[predictor.label]
positive_class = predictor.class_labels[1]

no_skill = len(y_true[y_true == positive_class]) / len(y_true)


model_groups = {
    "Group 1: Ensemble, NN(FastAI) & KNN": ["WeightedEnsemble_L2", "NeuralNetFastAI", "KNeighborsDist", "KNeighborsUnif"],
    "Group 2: Boosting Models (GBM/XGB)": ["LightGBMXT", "LightGBM", "LightGBMLarge", "XGBoost"],
    "Group 3: Trees & NN(Torch)": ["RandomForestEntr", "NeuralNetTorch", "ExtraTreesEntr", "RandomForestGini", "ExtraTreesGini"]
}


fig, axes = plt.subplots(1, 3, figsize=(24, 7))

for idx, (group_name, models) in enumerate(model_groups.items()):
    ax = axes[idx]
    
    # 绘制基准线 (No Skill)
    ax.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill', color='navy')
    
    for model_name in models:
        try:
            # 获取预测概率
            y_scores = predictor.predict_proba(test_data, model=model_name, as_pandas=True)[positive_class]
            
            # 计算 PR 曲线
            precision, recall, _ = precision_recall_curve(y_true, y_scores, pos_label=positive_class)
            # 计算 AP 
            ap_score = average_precision_score(y_true, y_scores, pos_label=positive_class)
            
            # 样式调整
            lw = 2.5 if model_name == "WeightedEnsemble_L2" else 1.5
            ax.plot(recall, precision, lw=lw, label=f'{model_name} (AP={ap_score:.4f})')
            
        except Exception:
            pass

    ax.set_xlabel('Recall (Sensitivity)')
    if idx == 0:
        ax.set_ylabel('Precision (PPV)')
    ax.set_title(f'PR Curve - {group_name}', fontsize=12, fontweight='bold')
    ax.legend(loc="best", fontsize=9)
    ax.grid(True, alpha=0.3)
    ax.set_ylim([0.0, 1.05])
    ax.set_xlim([0.0, 1.0])

    
    plt.savefig('PR.png', dpi=300, bbox_inches='tight')
    
plt.tight_layout()
plt.show()
```

```python
import numpy as np

def calculate_net_benefit(y_true, y_prob, thresholds):
    net_benefits = []
    n = len(y_true)
    
    for thresh in thresholds:
        # 在当前阈值下的预测结果
        y_pred = (y_prob >= thresh).astype(int)
        
        tp = np.sum((y_true == positive_class) & (y_pred == 1))
        fp = np.sum((y_true != positive_class) & (y_pred == 1))
        
        # 核心公式
        if thresh == 1.0:
            nb = 0 # 避免除以0
        else:
            weight = thresh / (1 - thresh)
            nb = (tp / n) - (fp / n) * weight
        net_benefits.append(nb)
        
    return np.array(net_benefits)


thresholds = np.linspace(0.01, 0.99, 100) # 生成 1% 到 99% 的阈值点
fig, axes = plt.subplots(1, 3, figsize=(24, 7))

total_pos = np.sum(y_true == positive_class)
total_neg = len(y_true) - total_pos
n = len(y_true)
treat_all = (total_pos / n) - (total_neg / n) * (thresholds / (1 - thresholds))
# Treat None 的收益永远是 0

for idx, (group_name, models) in enumerate(model_groups.items()):
    ax = axes[idx]
    

    ax.plot(thresholds, treat_all, linestyle=':', color='gray', label='Treat All')
    ax.axhline(y=0, color='black', linestyle='-', lw=1, label='Treat None')
    

    for model_name in models:
        try:
            y_scores = predictor.predict_proba(test_data, model=model_name, as_pandas=True)[positive_class]
         
            nb_model = calculate_net_benefit(y_true, y_scores, thresholds)
            
            lw = 2.5 if model_name == "WeightedEnsemble_L2" else 1.5
            ax.plot(thresholds, nb_model, lw=lw, label=model_name)
            
        except Exception:
            pass

    
    y_max = max(total_pos/n, np.max(treat_all[treat_all > -0.1])) + 0.05
    ax.set_ylim([-0.05, y_max]) 
    ax.set_xlim([0, 0.8]) # 阈值通常看 0 到 0.8 即可，太高通常没有意义
    
    ax.set_xlabel('Threshold Probability')
    if idx == 0:
        ax.set_ylabel('Net Benefit')
    ax.set_title(f'DCA - {group_name}', fontsize=12, fontweight='bold')
    ax.legend(loc="upper right", fontsize=9)
    ax.grid(True, alpha=0.3)

    
    plt.savefig('DCA.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
from sklearn.metrics import brier_score_loss


y_true = test_data[predictor.label]
positive_class = predictor.class_labels[1]


model_groups = {
    "Group 1: Ensemble, NN(FastAI) & KNN": ["WeightedEnsemble_L2", "NeuralNetFastAI", "KNeighborsDist", "KNeighborsUnif"],
    "Group 2: Boosting Models (GBM/XGB)": ["LightGBMXT", "LightGBM", "LightGBMLarge", "XGBoost"],
    "Group 3: Trees & NN(Torch)": ["RandomForestEntr", "NeuralNetTorch", "ExtraTreesEntr", "RandomForestGini", "ExtraTreesGini"]
}


fig, axes = plt.subplots(1, 3, figsize=(24, 8))

for idx, (group_name, models) in enumerate(model_groups.items()):
    ax = axes[idx]
    

    ax.plot([0, 1], [0, 1], "k:", linewidth=2, label="Perfectly Calibrated")
    
    for model_name in models:
        try:

            y_scores = predictor.predict_proba(test_data, model=model_name, as_pandas=True)[positive_class]
            

            prob_true, prob_pred = calibration_curve(y_true, y_scores, n_bins=10, strategy='uniform')

            clf_score = brier_score_loss(y_true, y_scores, pos_label=positive_class)
            
            lw = 2.5 if model_name == "WeightedEnsemble_L2" else 1.5
            marker = 'o' if len(prob_pred) < 15 else None 
            
            ax.plot(prob_pred, prob_true, marker=marker, linewidth=lw, 
                    label=f'{model_name} (Brier={clf_score:.4f})')
            
        except Exception:
            pass

    ax.set_xlabel('Mean Predicted Probability')
    if idx == 0:
        ax.set_ylabel('Fraction of Positives (Actual)')
    
    ax.set_ylim([-0.05, 1.05])
    ax.set_xlim([-0.05, 1.05])
    ax.set_title(f'Calibration Plot - {group_name}', fontsize=12, fontweight='bold')
    ax.legend(loc="best", fontsize=9)
    ax.grid(True, alpha=0.3)
    
    plt.savefig('CP.png', dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
```

```python
import shap
import numpy as np
import matplotlib.pyplot as plt


test_sample = test_data.drop(columns=['Migraine']).head(100) 
train_summary = train_data.drop(columns=['Migraine']).sample(100, random_state=42)


# -----------------------------------------------------------
# 修复后的预测函数包装器
# -----------------------------------------------------------
def autogluon_predict_proba(data):
    # 关键步骤：SHAP 内部会传入 numpy 数组，我们要把它转回 DataFrame
    if isinstance(data, np.ndarray):
        # 使用测试集的列名来重建 DataFrame
        data = pd.DataFrame(data, columns=test_sample.columns)
    
    # 返回正例 (Class 1) 的概率
    return predictor.predict_proba(data).iloc[:, 1].values

# -----------------------------------------------------------
# 重新运行计算
# -----------------------------------------------------------
print("正在尝试重新计算 SHAP 值...")
# 注意：KernelExplainer 对数据非常敏感，确保 background_data 也是 DataFrame
explainer = shap.KernelExplainer(autogluon_predict_proba, train_summary)
shap_values = explainer.shap_values(test_sample)

print("计算成功！")


print("正在计算 SHAP 值，请稍候...")

explainer = shap.KernelExplainer(autogluon_predict_proba, train_summary)
# 计算测试集样本的 SHAP 值
shap_values = explainer.shap_values(test_sample)


plt.figure(figsize=(12, 10))
shap.summary_plot(shap_values, test_sample, show=False)
plt.title("SHAP Interpretation for Migraine Prediction (Best Model)", fontsize=16)
plt.tight_layout()
plt.savefig("shap_beeswarm_plot.png", dpi=300)
plt.show()

print("SHAP 蜂群图已保存为 'shap_beeswarm_plot.png'")


plt.figure(figsize=(12, 10))
shap.summary_plot(shap_values, test_sample, plot_type="bar", show=False)
plt.title("Global Feature Importance (SHAP)", fontsize=16)
plt.tight_layout()
plt.savefig("shap_bar_plot.png", dpi=300)
plt.show()
```

```python
explanation = shap.Explanation(
    values=shap_values, 
    base_values=explainer.expected_value, 
    data=test_sample.values, 
    feature_names=test_sample.columns
)


plt.figure(figsize=(12, 8))
# instance_order 相似度排序
shap.plots.heatmap(explanation, show=False)
plt.title("SHAP Heatmap for Migraine Prediction", fontsize=16)
plt.tight_layout()
plt.savefig("shap_heatmap_highres.png", dpi=300, bbox_inches='tight')
plt.show()
```
